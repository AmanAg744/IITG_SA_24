### 6. Using seaborn make different types of plots, observe the trend and answer the questions given in the form.

```python
sns.scatterplot
```

```python
sns.scatterplot(x='Attack', y='Defense', data=pokemon, hue='Legendary')
plt.title('Attack vs. Defense')
plt.show()

```

```python
sns.boxplot(x='Generation', y='Total', data=pokemon)
plt.title('Total Stats Distribution by Generation')
plt.show()

```

### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
gen4ghost = pokemon[((pokemon['Type 1'] == 'Ghost') | (pokemon['Type 2'] == 'Ghost')) & (pokemon['Generation'] == 4) & (pokemon['Legendary'] == False)]
gen4ghost_sorted = gen4ghost.sort_values(by = 'Speed',ascending = False)
gen4ghost_sorted.iloc[1]
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
charizard = pokemon[pokemon['Name'] == 'Charizard']
charizard_attack = charizard['Attack'].values[0]
charizard_defence = charizard['Defense'].values[0]
numberpok = pokemon[(pokemon['Attack'] < charizard_attack) & (pokemon['Defense'] > charizard_defence) & (pokemon['Legendary'] == False)].shape[0]
numberpok
```
